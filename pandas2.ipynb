***Group By  and Aggregating***

```python
# df.head()
```

```python
# df.reset_index(inplace=True)
```

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\arshi\Downloads\Desktop\Brototype\pandas\Flavors.csv")
df.head()
```

```python
pd.pivot_table(df,columns='Flavor',values=['Flavor Rating','Texture Rating','Total Rating'],aggfunc='mean')
```

```python
gr_by_frame=df.groupby('Flavor').mean('Flavor Rating','Texture Rating','Total Rating')
gr_by_frame.sort_values(by='Total Rating',ascending=False)
```

```python
df.groupby('Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum(['Flavor Rating','Texture Rating','Total Rating'])
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','count','max','min'],'Total Rating':['mean','count','max','min'],'Texture Rating':['mean','count','max','min']})
```

```python
df.groupby(['Base Flavor','Liked']).max()
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','count','max','min'],'Total Rating':['mean','count','max','min'],'Texture Rating':['mean','count','max','min']})
```

```python
df.groupby('Base Flavor').describe()
```

***Merge Concatenate Join***

```python
df1=pd.read_csv(r"C:\Users\arshi\Downloads\Desktop\Brototype\pandas\LOTR.csv")
df1.head()
```

```python
df2=pd.read_csv(r"C:\Users\arshi\Downloads\Desktop\Brototype\pandas\LOTR 2.csv")
df2.head()
```

```python
# Merge
df1.merge(df2) # it will do as inner join
```

```python
# inner join
df1.merge(df2,how='inner',on =['FellowshipID','FirstName'])
```

```python
#  outer join
df1.merge(df2,how='outer')
```

```python
#  left join
df1.merge(df2,how='left')
```

```python
# right join
df1.merge(df2,how='right')
```

```python
# cross join
df1.merge(df2,how='cross')
```

**Join**

```python
df1.join(df2)
```

```python
df1.join(df2,on='FellowshipID', how='outer',lsuffix='_l',rsuffix='_r')
```

```python
df4=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_le',rsuffix='_ri')
df4
```

**Concatenate**

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='outer')
```

```python
pd.concat([df1,df2],join='inner')
```
