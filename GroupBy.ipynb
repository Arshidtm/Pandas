```python
import pandas as pd
```

```python
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 'Appliances'],
    'Sales': [1000, 500, 1500, 750, 900],
    'Date': ['2024-01-01', '2024-02-15', '2024-03-10', '2024-04-05', '2024-05-12']
}

df = pd.DataFrame(data)
df
```

```python
categories=df.groupby('Category')['Sales'].sum().reset_index()
categories
```

```python
categories.columns=['Category','Total Sales']
categories
```

```python
Add a new column to the DataFrame that indicates whether the sales for that row are above or below the average sales of the corresponding category.
```

```python
highest=categories.loc[categories['Total Sales'].idxmax()]
highest
```

```python

df['Avg']=df.groupby('Category')['Sales'].transform('mean')
df
```

```python
df['Avg_above_below']=df.apply(lambda row : 'above' if row['Sales'] > row['Avg'] else 'Below',axis=1)
df
```

```python
import pandas as pd

data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Department': ['HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'],
    'Salary': [60000, 70000, 90000, 55000, 65000, 80000, 50000],
    'Joining Date': ['2022-01-15', '2021-07-30', '2020-06-25', '2022-03-10', '2021-11-12', '2019-08-20', '2023-01-05']
}

df = pd.DataFrame(data)
df
```

```python
sal_dep=df.groupby('Department')['Salary'].mean().reset_index()
sal_dep.columns=['Department','Avg_salary']
sal_dep
```

```python
sal_dep.loc[sal_dep['Avg_salary'].idxmin()]
```

```python
df['avg_sal']=df.groupby('Department')['Salary'].transform('mean')
df
```

```python
df['Sal_above_below']=df.apply(lambda row: 'above' if row['Salary']>=row['avg_sal'] else 'below',axis=1)
df
```

```python
data = {
    'Product': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'East', 'East', 'East'],
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450, 500],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09'],
    'Discount': [5, 10, 0, 5, 10, 15, 5, 0, 10]  # percentage discounts
}

df = pd.DataFrame(data)
df
```

```python
avg_total=df.groupby(['Region','Product']).agg(Total_sales=('Sales','sum'),Average_sales=('Sales','mean')).reset_index()
avg_total
```

```python
df['Revenue']=df['Sales']*(1-df['Discount']/100)
```

```python
df
```

```python
avg_rev=df.groupby('Region')['Revenue'].mean().reset_index()
avg_rev
```

```python
avg_rev.loc[avg_rev['Revenue'].idxmax()]
```

```python
df['Avg_sales']=df.groupby('Product')['Sales'].transform('mean')
df
```

```python
df['Sales_above']=df.apply(lambda row : 'above' if row['Sales']>=row['Avg_sales'] else 'below',axis=1)
df
```

```python
import numpy as np

data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles', 'Chicago'],
    'Date': pd.date_range(start='2024-01-01', periods=9, freq='D'),
    'Temperature': [30, 25, 28, 32, 26, 29, 31, 27, 30],
    'Humidity': [70, 65, 75, 72, 67, 78, 74, 68, 76],
    'Wind Speed': [10, 8, 12, 9, 7, 11, 10, 8, 13]
}

df = pd.DataFrame(data)
df
```
