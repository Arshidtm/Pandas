```python
import pandas as pd
```

```python
df1=pd.DataFrame({
    "id":[1,2,3,4,5,6],
    'Name':['Arun','Raju','Rohan','Sonu','Manu','Ram'],
    'Age':[21,22,23,21,19,24],
    'Dept_id':[100,101,100,102,109,103]
})
df1
```

```python
df2=pd.DataFrame({
    
    'depName':['ML','MERN','MEAN','CS'],
    
    'Dept_id':[100,101,102,104]
})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
df3=pd.concat((df1,df2),axis=1)
df3
```

```python
df1[df1['Age']>21]
```

```python
df3[(df3['Age']>20) & (df3['depName']=='ML')]
```

```python
df = pd.DataFrame([[1, 2], [3, 4]])
result = df.apply(lambda x: x ** 2,axis=0)
print(result)  # Output: DataFrame with each element squared
```

```python
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
        'Value': [10, 20, 30, 40, 50, 60]}

df = pd.DataFrame(data)
df
```

```python
df.groupby('Category').sum()
```

```python
df.groupby('Category').mean()
```

```python
df.groupby('Category').count()
```

```python
df.groupby('Category').agg({'Value':['min','max','count','sum','mean']})
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'C', 'C'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'X', 'Y'],
    'Value': [10, 20, 30, 40, 50, 60, 70]
}
df=pd.DataFrame(data)
```

```python
df.groupby(['Category','Subcategory']).sum()
```

```python
df.pivot_table(index='Category',values='Value',columns='Subcategory',aggfunc=['describe'])
```

```python
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-01'],
    'City': ['A', 'B', 'A'],
    'Temperature': [30, 32, 28]
}

df = pd.DataFrame(data)
df=df.drop_duplicates(subset=['Date','City'])

# This will raise an error due to duplicate entries for 'Date' and 'City'
pivot = df.pivot(index='Date', columns='City', values='Temperature')
pivot
```
